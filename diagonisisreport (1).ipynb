```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
import warnings
import joblib  # Import joblib for saving the model

warnings.filterwarnings("ignore")
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,ConfusionMatrixDisplay, \
                            precision_score, recall_score, f1_score, roc_auc_score,roc_curve
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
import numpy as np

# Define the models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    }

# Stratified K-Fold Cross Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Iterate through the models
for model_name, model in models.items():
    print(f"Model: {model_name}")

    # Cross-validation for accuracy
    cv_accuracy = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    cv_f1 = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1_weighted')
    cv_precision = cross_val_score(model, X_train, y_train, cv=cv, scoring='precision_weighted')
    cv_recall = cross_val_score(model, X_train, y_train, cv=cv, scoring='recall_weighted')

    # Print Cross-Validation Results
    print("Cross-Validation Results:")
    print("- Accuracy: {:.4f}  {:.4f}".format(cv_accuracy.mean(), cv_accuracy.std()))
    print("- F1 Score: {:.4f}  {:.4f}".format(cv_f1.mean(), cv_f1.std()))
    print("- Precision: {:.4f}  {:.4f}".format(cv_precision.mean(), cv_precision.std()))
    print("- Recall: {:.4f}  {:.4f}".format(cv_recall.mean(), cv_recall.std()))

    # Fit the model and evaluate test set performance
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred, average='weighted')
    test_precision = precision_score(y_test, y_test_pred, average='weighted')
    test_recall = recall_score(y_test, y_test_pred, average='weighted')

    print("\nTest Set Performance:")
    print("- Accuracy: {:.4f}".format(test_accuracy))
    print("- F1 Score: {:.4f}".format(test_f1))
    print("- Precision: {:.4f}".format(test_precision))
    print("- Recall: {:.4f}".format(test_recall))

    print("="*40)
```

```python
import joblib
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# ========================
# Save Encoders and Model
# ========================

# Encoder for 'Risk Category'
encoder_risk = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder_risk.fit(df[['Risk Category']])
joblib.dump(encoder_risk, 'risk_encoder.pkl')

# Encoder for 'Gender'
encoder_gender = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
df['Gender'] = df['Gender'].astype(str).str.lower()  # Normalize during training
encoder_gender.fit(df[['Gender']])
joblib.dump(encoder_gender, 'gender_encoder.pkl')

# Train and save the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
joblib.dump(rf_model, 'random_forest_model.pkl')

print("Model and encoders saved successfully.")

# ========================================
# Load the model and test with new data
# ========================================

# Load saved model and encoders
loaded_rf_model = joblib.load('random_forest_model.pkl')
loaded_risk_encoder = joblib.load('risk_encoder.pkl')
loaded_gender_encoder = joblib.load('gender_encoder.pkl')

# Example new data for prediction
new_data = pd.DataFrame({
    'Heart Rate': [72],
    'Respiratory Rate': [16],
    'Body Temperature': [36.8],
    'Oxygen Saturation': [98],
    'Systolic Blood Pressure': [120],
    'Diastolic Blood Pressure': [80],
    'Age': [29],
    'Gender': ['Male'],  # Categorical column
    'Derived_HRV': [50],
    'Derived_Pulse_Pressure': [40],
    'Derived_BMI': [23.5],
    'Derived_MAP': [93]
})

# =======================================
# Preprocess new data (MATCH TRAINING)
# =======================================

# Normalize Gender
new_data['Gender'] = new_data['Gender'].str.lower()

# Encode 'Gender'
encoded_gender = loaded_gender_encoder.transform(new_data[['Gender']])

# Drop 'Gender' from new_data and append encoded columns
new_data = new_data.drop(['Gender'], axis=1)
encoded_gender_df = pd.DataFrame(encoded_gender, columns=loaded_gender_encoder.get_feature_names_out(['Gender']))
new_data = pd.concat([new_data.reset_index(drop=True), encoded_gender_df], axis=1)

# Ensure feature order matches training
required_columns = X_train.columns  # Columns used for training
new_data = new_data.reindex(columns=required_columns, fill_value=0)  # Fill missing columns with 0


```

```python
# Print the categories used by the encoder
print("Encoder categories:", loaded_risk_encoder.categories_)
```

```python
# =======================================
# Manually map predicted index to original labels
# =======================================

# Predicted risk category as numeric index (0.0 or 1.0)
prediction = loaded_rf_model.predict(new_data)

# Manually define a mapping from encoded labels to the original risk categories
# Assuming the encoder categories correspond to ['Low', 'High']
category_mapping = {0.0: 'Low Risk' , 1.0: 'High Risk'}

# Convert the predicted numeric value to the original category label
predicted_risk_category_index = float(prediction[0])  # Ensure it's a float
predicted_risk = category_mapping[predicted_risk_category_index]

print("Predicted Risk Category:", predicted_risk)
```
